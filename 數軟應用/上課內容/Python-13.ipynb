### 新增一個套件的方法（到終端機去裝以外的方式）

```python
!pip install googletrans
```

```python
from googletrans import Translator
```

```python
respond_robot=Translator()
```

```python
result=respond_robot.translate("Good Morning my dear!",dest="zh-tw")
```

```python
print(result.text)
```

#### 基本指令
*'src'：原文
*'dest'：目標語言

常用語言：
*中文：zh-tw
*日文：ja
*英文：en
