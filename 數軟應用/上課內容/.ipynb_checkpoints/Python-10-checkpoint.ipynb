```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## 模擬版資料做SVM

```python
x=np.array([[6,6],[10,8],[3,7],[8,7]])
```

```python
x
```

```python
x[:,0]
```

```python
plt.scatter(x[:,0],x[:,1])
```

```python
y=np.array([1,2,1,2])
```

```python
plt.scatter(x[:,0],x[:,1],c=y,cmap='Paired',s=100) ##顏色由y決定,camp是改顏色的地圖,s=size
```

### Step 1.打開函數學習機

```python
from sklearn.svm import SVC
```

```python
clf=SVC()
```

### Step 2.學習

```python
clf.fit(x,y)
```

### Step 3.預測

```python
y_predict=clf.predict(x)
```

```python
y_predict
```

```python
y
```

```python
xr=np.random.rand(100)*7+3
yr=np.random.rand(100)*2+6
```

```python
plt.scatter(xr,yr)
```

```python
print('X',xr[:5])
print('Y',yr[:5])
```

```python
X=np.c_[xr,yr]
```

```python
Y=clf.predict(X)
```

```python
plt.scatter(x[:,0],x[:,1],c=y,cmap='Paired',s=150)
plt.scatter(X[:,0],X[:,1],c=Y,cmap='Paired')
```

### 結果呈現(another way)

```python
x_list=np.linspace(3,10,50)
y_list=np.linspace(6,8,50)
X,Y=np.meshgrid(x_list,y_list)
P=np.c_[X.ravel(),Y.ravel()] ##ravel用來拉平
C=clf.predict(P)
plt.scatter(P[:,0],P[:,1],c=C)
```

### 插播:contour用法(等高線)

```python
xL=np.array([1,2,3,4])
yL=np.array([5,6,7,8])
X,Y=np.meshgrid(xL,yL)
```

```python
X
```

```python
Z=np.random.randint(1,3,(4,4))
plt.contour(X,Y,Z)
plt.scatter(X.ravel(),Y.ravel(),c=Z.ravel())
```

```python
Z=np.random.randint(1,3,(4,4))
plt.contourf(X,Y,Z) ##contour[f]
plt.scatter(X.ravel(),Y.ravel(),c=Z.ravel())
```

### 資料視覺化終極版

```python
x_arr=np.arange(3,10,0.02)
y_arr=np.arange(6,8,0.02)
X,Y=np.meshgrid(x_arr,y_arr)
P=np.c_[X.ravel(),Y.ravel()]
z=clf.predict(P)
Z=z.reshape(X.shape)

plt.contourf(X,Y,Z,cmap='Paired',alpha=0.3)
plt.scatter(x[:,0],x[:,1],c=y,cmap='Paired',s=150)
```

## 鳶尾花數據庫

```python
from sklearn.datasets import load_iris
```

```python
iris=load_iris()
```

```python
print(iris.DESCR)
```

```python
x=iris.data #輸入
y=iris.target #輸出
```

```python
x[:5]
```

```python
y[:5] #都是第0種花
```

```python
X=x[:,:2]
```

```python
Y=y
```

```python
len(X)
```

## 區分訓練資料、測試資料

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.2,random_state=0)#最後一個表每次選會一樣
```

```python
plt.scatter(X[:,0],X[:,1],c=Y,cmap='Paired')
```

### Step 1.開一台函數學習機

```python
clf=SVC()
```

### Step 2.訓練

```python
clf.fit(x_train,y_train)
```

### Step 3.預測

```python
y_predict=clf.predict(x_test)
```

```python
y_predict
```

```python
y_test
```

```python
y_predict -y_test
```

```python
plt.scatter(x_test[:,0],x_test[:,1],c=y_predict-y_test)
```

```python
x0=np.arange(3.8,8.2,0.02)
y0=np.arange(1.8,4.7,0.02)

xm,ym=np.meshgrid(x0,y0)
P=np.c_[xm.ravel(),ym.ravel()]
z=clf.predict(P)
Z=z.reshape(xm.shape)
plt.contourf(xm,ym,Z,cmap='Paired',alpha=0.2)
plt.scatter(x_test[:,0],x_test[:,1],cmap='Paired',c=y_test)
```
