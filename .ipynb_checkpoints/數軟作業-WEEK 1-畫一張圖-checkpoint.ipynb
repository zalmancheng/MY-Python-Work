```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

x=np.linspace(-14,14,400)
y=np.tan(x)
y[y>5]=np.inf
y[y<-5]=-np.inf

plt.xkcd()
plt.scatter(x,y)
```
