# 108學測缺考人數總表（大考中心）

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df=pd.read_excel('People.xls')
```

# 計算並插入各科缺考比例

```python
df.drop('英文缺考比例',axis=1,inplace=True)
```

```python
df.drop('英文缺考比例',axis=1,inplace=True)
```

```python
df.insert(3,'英文缺考比例',(df.英文缺考/df.英文實到)*100)
df.insert(6,'國文缺考比例',(df.國文缺考/df.國文實到)*100)
df.insert(9,'社會缺考比例',(df.社會缺考/df.社會實到)*100)
df.insert(12,'數學缺考比例',(df.數學缺考/df.數學實到)*100)
df.insert(15,'國寫缺考比例',(df.國寫缺考/df.國寫實到)*100)
df.insert(18,'自然缺考比例',(df.自然缺考/df.自然實到)*100)
```

```python
df.head()
```

# 計算平均缺考比例並排序

```python
df['平均缺考比例']=(df.英文缺考比例+df.國文缺考比例+df.社會缺考比例+df.數學缺考比例+df.國寫缺考比例+df.自然缺考比例)/6
```

```python
df.sort_values(by=["平均缺考比例"], ascending=False)
```

```python
df.head()
```

### 缺考情況最嚴重前五名

```python
df.sort_values(by=["平均缺考比例"], ascending=False).head()
```

### 缺考情況最輕微前五名

```python
df.sort_values(by=["平均缺考比例"], ascending=True).head()
```

# 各科缺考比例分佈圖

```python
df.英文缺考比例.plot(color='b'),df.國文缺考比例.plot(color='y'),df.社會缺考比例.plot(color='g'),df.數學缺考比例.plot(color='k'),df.國寫缺考比例.plot(color='m'),df.自然缺考比例.plot(color='c');
```
